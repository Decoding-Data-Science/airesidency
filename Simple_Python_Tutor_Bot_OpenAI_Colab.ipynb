```python
from google.colab import userdata
from openai import OpenAI
import os

# Load API key from Google Colab Secrets
os.environ["OPENAI_API_KEY"] = userdata.get("openai")

# Create OpenAI client
client = OpenAI()
```

## Step 3: Build a simple Python Tutor Bot

This is the main demo.

The user types a Python question, and the model responds like a friendly Python tutor.


```python
# Simple Python Tutor Bot with geenrating models

system_prompt = """

Answer only beginner-friendly, Python-related questions.
If a question is not about Python, politely decline and state that you only answer Python questions.

For any Python question:
- Ask follow-up questions if clarification is needed.
- Always provide a simple, readable code snippet, fully commented.
- Use analogies to explain concepts, tailored for beginners.
- Never answer non-Python questions.

**Output Format:**
Respond in a clear, friendly tone.
Start with a brief analogy or explanation.
Then provide a fully-commented code snippet, with each logical step explained.
For non-Python topics, respond only with:
"I'm here to help with Python questions only. Please ask a Python-related question."

**Example 1:**
Input: How can I add two numbers in Python?
Output:
Analogy: Adding two numbers in Python is like putting two apples together in a basket—you simply count them together.
Code:
# Adding two numbers in Python
a = 2  # first number
b = 3  # second number
sum = a + b  # add the numbers together
print(sum)  # Output will be 5

**Example 2:**
Input: How do I solve for x in algebra?
Output:
I'm here to help with Python questions only. Please ask a Python-related question.

**Example 3:**
Input: How does a for loop work in Python?
Output:
Analogy: A for loop in Python is like checking every book on a shelf, one at a time.
Code:
# Using a for loop in Python
books = ["book1", "book2", "book3"]  # a list of books
for book in books:  # go through each book, one by one
    print(book)  # print the current book

---

**Remember:**
- Only answer Python-related questions.
- Always give follow-up questions if a query is unclear.
- Include beginner-friendly analogies and clear, commented code.
- Decline non-Python questions politely.

**IMPORTANT INSTRUCTION REMINDER:**
Always answer Python-only questions with beginner-focused analogies and code. Politely decline all other topics.

"""

user_question = input("Ask your Python question: ")

response = client.chat.completions.create(
    model="gpt-4.1-mini",
    messages=[
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_question}
    ],
    temperature=0.4,
    top_p=0.2,
    max_completion_tokens=500
)

print(response.choices[0].message.content)
```

## Step 4: Change temperature and observe the difference

Temperature controls how focused or creative the answer is.

- Lower temperature: more focused and predictable
- Higher temperature: more creative and varied


```python
# Try the same Python question with a lower temperature

system_prompt = """
You are a friendly Python tutor for beginners.
Explain concepts clearly and use small examples.
"""

user_question = input("Ask your Python question: ")

response = client.chat.completions.create(
    model="gpt-4.1-mini",
    messages=[
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_question}
    ],
    temperature=1,
    top_p=0.2,
    max_completion_tokens=500
)

print(response.choices[0].message.content)
```

```python
# Try the same Python question with a higher temperature

system_prompt = """
You are a friendly Python tutor for beginners.
Explain concepts clearly and use small examples.
"""

user_question = input("Ask your Python question: ")

response = client.chat.completions.create(
    model="gpt-4.1-mini",
    messages=[
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_question}
    ],
    temperature=0.9,
    top_p=1.0,
    max_completion_tokens=500
)

print(response.choices[0].message.content)
```

## Simple explanation for participants

Use this explanation during the workshop:

- The **system prompt** defines the role of the assistant.
- The **user input** is the learner's question.
- The **model** decides which OpenAI model to use.
- The **temperature** controls how focused or creative the answer is.
- The **top_p** controls the model's word-choice range.
- The **max_completion_tokens** controls the maximum answer length.

For the first demo, keep it simple. Later, you can explain reasoning models, RAG, agents, tools, and more advanced configuration.


```python
# Simple Python Tutor Bot with reasoning model

system_prompt = """
You are a friendly Python tutor for beginners.

Your job:
- Explain Python concepts in simple language
- Use small code examples
- Avoid advanced jargon
- If the student makes a mistake, correct it gently
- Give step-by-step explanations
"""

user_question = input("Ask your Python question: ")

response = client.responses.create(
    model="gpt-5.5",
    instructions=system_prompt,
    input=user_question,
    reasoning={"effort": "high"},
    max_output_tokens=500
)

print(response.output_text)
```
